```python
import collections.abc
collections.Mapping = collections.abc.Mapping
from experta import *
class HeartExpert(KnowledgeEngine):
    def init(self):
        super().init()
        self.ones = 0
        self.zeros=0
    @Rule(Fact(age=P(lambda x: x > 55))&Fact(chol=P(lambda x: x > 240)))
    def high_risk_age_chol(self):
        print("High risk! Consult a doctor.(high age & high chol)")
        self.ones+= 1
    @Rule(Fact(trestbps=P(lambda x:x >140))& Fact(exang=1))
    def high_risk_bp_exang(self):
        print("High risk! Consult a doctor.(high trestbps & exang=1)")
        self.ones+= 1
    @Rule(Fact(thalach=P(lambda x: x <120))&Fact(oldpeak=P(lambda x: x > 2)))
    def high_risk_thalach_oldpeak(self):
        print("High risk! Consult a doctor.(low thalach & high oldpeak)")
        self.ones+= 1
    @Rule(Fact(ca=P(lambda x: x >=2))&Fact(thal=3))
    def high_risk_cp_thal(self):
        print("High risk! Consult a doctor.(high ca & thal=3)")
        self.ones+= 1
    @Rule(Fact(cp=0)&Fact(exang=1))
    def high_risk_cp_exang(self):
        print("High risk! Consult a doctor.(cp=0 & exang=1)")
        self.ones+= 1
    @Rule(Fact(age=P(lambda x: 45 <= x <= 55)) &Fact(chol=P(lambda x: 200 <= x <= 240)))
    def medium_risk_age_chol(self):
        print("Medium risk! Be careful.(age between 45 and 55 & chol between 200 and 240)")
        self.ones+= 1
    @Rule(Fact(trestbps=P(lambda x: 120 <= x <= 140)))
    def medium_risk_bp(self):
        print("Medium risk! Be careful.(trestpbs between 120 and 140)")
        self.ones+= 1
    @Rule(Fact(thalach=P(lambda x: 120 <= x <= 150)))
    def medium_risk_thalach(self):
        print("Medium risk! Be careful.(thalach between 120 and 150)")
        self.ones+= 1
    @Rule(Fact(oldpeak=P(lambda x: 1 <= x <= 2)))
    def medium_risk_oldpeak(self):
        print("Medium risk! Be careful.(oldpeak between 1 and 2)") 
        self.ones+= 1
    @Rule(Fact(age=P(lambda x: x < 45)) & Fact(chol=P(lambda x: x < 200)))
    def low_risk_age_chol(self):
        print("Low risk! Stay healthy.(age low & chol low)")
        self.zeros+= 1
    @Rule(Fact(thalach=P(lambda x: x > 150)) & Fact(exang=0))
    def low_risk_thalach(self):
        print("Low risk! Stay healthy.(thalach low & exang=0)")
        self.zeros+= 1
    @Rule(Fact(cp=P(lambda x: x in [2, 3])))
    def low_risk_cp(self):
        print("Low risk! Stay healthy.(cp high)")
        self.zeros+= 1
    @Rule(Fact(ca=0) & Fact(thal=2))
    def low_risk_ca_thal(self):
        print("Low risk! Stay healthy.(ca=0 & thal=2)")
        self.zeros+= 1
engine = HeartExpert()
engine.reset()
 age = int(input("Enter age: "))
while True:
    gender = input("Enter gender(female,male): ").strip().lower()
    if gender=="male":
        sex=0
    elif gender=="female":
        sex=1
    if sex in range (0,2):
        break
    print("Invalid input! Enter female or male only.")
while True:
    chest_pain = input("Chest pain level (no pain,low,medium,high)): ").strip().lower()
    if chest_pain=="no pain":
        cp=0
    elif chest_pain=="low":
        cp=1
    elif chest_pain=='medium':
        cp=2
    elif chest_pain=="high":
        cp=4
    if cp in range(0,4):
        break
    print("Invalid input! Enter (no pain,low,medium,high) only.")
trestbps = int(input("Resting blood pressure: "))
chol = int(input("Cholesterol level: "))
while True:
    fasting = input("Fasting blood sugar (yes, no): ").strip().lower()
    if fasting=="yes":
        fbs=1
    elif fasting=='no':
        fbs=0
    if fbs in range(0,2):
        break
    print("Invalid input! Enter yes or no only.")
restecg = int(input("Rest ECG result (0~2): "))
thalach = int(input("Max heart rate achieved: "))
```
