```python
age = int(input("Enter age: "))
while True:
    gender = input("Enter gender(female,male): ").strip().lower()
    if gender=="male":
        sex=0
    elif gender=="female":
        sex=1
    if sex in range (0,2):
        break
    print("Invalid input! Enter female or male only.")
while True:
    chest_pain = input("Chest pain level (no pain,low,medium,high)): ").strip().lower()
    if chest_pain=="no pain":
        cp=0
    elif chest_pain=="low":
        cp=1
    elif chest_pain=='medium':
        cp=2
    elif chest_pain=="high":
        cp=4
    if cp in range(0,4):
        break
    print("Invalid input! Enter (no pain,low,medium,high) only.")
trestbps = int(input("Resting blood pressure: "))
chol = int(input("Cholesterol level: "))
while True:
    fasting = input("Fasting blood sugar (yes, no): ").strip().lower()
    if fasting=="yes":
        fbs=1
    elif fasting=='no':
        fbs=0
    if fbs in range(0,2):
        break
    print("Invalid input! Enter no or no only.")
restecg = int(input("Rest ECG result (0~2): "))
thalach = int(input("Max heart rate achieved: "))
while True:
    exercise = input("Exercise induced angina (yes,no): ").strip().lower()
    if exercise=="yes":
        exang=1
    elif exercise=="no":
        exang=0
    if exang in range(0,2):
        break
    print("Invalid input! Enter yes or no only.")
oldpeak = float(input("Oldpeak value: "))
slope = int(input("Slope (0~2): "))
ca = int(input("Number of vessels (0~3): "))
thal = int(input("Thal value (0~3): "))
patient= {
        "age": age,
        "sex": sex,
        "cp": cp,
        "trestbps": trestbps,
        "chol": chol,
        "fbs": fbs,
        "restecg": restecg,
        "thalach": thalach,
        "exang": exang,
        "oldpeak": oldpeak,
        "slope": slope,
        "ca": ca,
        "thal": thal,
    }
engine = HeartExpert()
engine.reset()
for key, value in patient.items():
    engine.declare(Fact(**{key: value}))
engine.run()
if engine.result == 1:
    print("Prediction: Heart Disease")
elif engine.result == 0:
    print("Prediction: No Heart Disease")
else:
    print("No decision made")
```
