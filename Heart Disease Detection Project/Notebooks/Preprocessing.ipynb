```python
#Step 1 : Data processing
## LOAD DATASET :
import numpy as np
import pandas as pd
data= pd.read_csv("D:/Games/heart.csv")
data.head()
## HANDLING MISSING VALUES :
print("null number:",data.isnull().sum())
print("dupilcated number:",data.duplicated().sum())
data["restecg"]=data["restecg"].fillna(data["restecg"].mode()[0])
data["oldpeak"]=data["oldpeak"].fillna(data["oldpeak"].mean())
data.drop_duplicates(inplace=True) 
data
from sklearn.preprocessing import MinMaxScaler
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]
scaler = MinMaxScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
data
cols=["cp","restecg","slope","ca","thal"]
data_encoded=pd.get_dummies(data,columns=cols)
print(data_encoded.head().to_string(index=False)) # CONVERT DATA FRAME INTO CLEAN TABLE
##CORRELATION:
correlation=data.corr()["target"]
selected = correlation[abs(correlation) > 0.3]  # abs= ENSURE THAT CODE FOCUS ON COLERATION STRENGTH NOT ITS DIRECTION
print(selected)
## FEATURE SELECTION : 
features=selected.index
features = features.drop("target")
data_selected = data[features]
data_selected
## SAVE CLEANED DATA :
data.to_csv("cleaned_data.csv", index=False)
```
