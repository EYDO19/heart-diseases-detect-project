```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

tree_model = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=5,
    random_state=42
)

tree_model.fit(X_train, y_train)

y_pred_tree = tree_model.predict(X_test)

accuracy_tree = accuracy_score(y_test, y_pred_tree)
precision_tree = precision_score(y_test, y_pred_tree)
recall_tree = recall_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree)

print("Decision Tree Results")
print("Accuracy:", accuracy_tree*100,"%")
print("Precision:", precision_tree*100,"%")
print("Recall:", recall_tree*100,"%")
print("F1 Score:", f1_tree*100,"%")


def expert_system(row):

    risk=0

    if row["age"]>55:
        risk+=1
    if row["chol"]>240:
        risk+=1
    if row["trestbps"]>140:
        risk+=1
    if row["fbs"]==1:
        risk+=1
    if row["exang"]==1:
        risk+=1
    if row["thalach"]<120:
        risk+=1
    if row["oldpeak"]>2:
        risk+=1
    return 1 if risk>=2 else 0
y_pred_expert = X_test.apply(expert_system, axis=1)

accuracy_expert = accuracy_score(y_test, y_pred_expert)
precision_expert = precision_score(y_test, y_pred_expert, zero_division=0)
recall_expert = recall_score(y_test, y_pred_expert, zero_division=0)
f1_expert = f1_score(y_test, y_pred_expert, zero_division=0)


print("\nExpert System Results")
print("Accuracy:", accuracy_expert*100,"%")
print("Precision:", precision_expert*100,"%")
print("Recall:", recall_expert*100,"%")
print("F1 Score:", f1_expert*100,"%")
print("\nModel Comparison")
print("Decision Tree Accuracy:", accuracy_tree*100,"%")
print("Expert System Accuracy:", accuracy_expert*100,"%")
```
